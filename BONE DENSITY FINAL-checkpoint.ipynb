# Loading and Preprocessing the DICOM Images

```python
import os
import numpy as np
import pydicom
import cv2
import matplotlib.pyplot as plt
```

```python
def load_dicom_image(file_path):
    dicom = pydicom.dcmread(file_path)
    image = dicom.pixel_array
    image = cv2.resize(image, (224, 224))  # Resize to 224x224 for DenseNet
    image = image.astype(np.float32)
    image = (image - np.min(image)) / (np.max(image) - np.min(image))  # Normalize
    image = np.stack((image,)*3, axis=-1)  # Convert to 3-channel
    return image
```

```python
# Load all images from the directory
dataset_dir = r"C:\Harsh\Bone Density Internship\Dataset\dicom_dir"
images = []
for filename in os.listdir(dataset_dir):
    if filename.endswith(".dcm"):
        file_path = os.path.join(dataset_dir, filename)
        images.append(load_dicom_image(file_path))

images = np.array(images)
print(f"Loaded {len(images)} DICOM images.")
```

# Displaying some sample images after loading and preprocessing

```python
def show_sample_images(images, title):
    plt.figure(figsize=(10, 10))
    for i in range(9):
        plt.subplot(3, 3, i + 1)
        plt.imshow(images[i], cmap='gray')
        plt.title(f"Image {i + 1}")
        plt.axis('off')
    plt.suptitle(title)
    plt.show()

show_sample_images(images, "Sample DICOM Images After Preprocessing")
```

# Using a Pre-trained Model for Feature Extraction

```python
from tensorflow.keras.applications import DenseNet121
from tensorflow.keras.models import Model
```

```python
# Load the pre-trained DenseNet121 model without the top classification layer
base_model = DenseNet121(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
```

```python
# Create a new model that outputs the features from the last convolutional block
model = Model(inputs=base_model.input, outputs=base_model.output)
```

```python
# Extract features from the images
features = model.predict(images)
print(f"Extracted features shape: {features.shape}")
```

# Clustering the Images

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
# Flatten the features for clustering
flattened_features = features.reshape(features.shape[0], -1)
```

```python
# Perform PCA to reduce dimensionality (optional but helps in visualization)
pca = PCA(n_components=50)
pca_features = pca.fit_transform(flattened_features)
```

```python
# Apply K-means clustering
n_clusters = 3  # Assuming we want 3 clusters for Normal, Osteopenia, and Osteoporosis
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
clusters = kmeans.fit_predict(pca_features)
```

```python
# Plot the clusters
plt.scatter(pca_features[:, 0], pca_features[:, 1], c=clusters, cmap='viridis')
plt.title('Clusters of DICOM Images')
plt.show()
```

# Training a Classifier with Pseudo-Labels

```python
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, Input ,Dropout
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
from sklearn.metrics import confusion_matrix, f1_score, accuracy_score, precision_score, recall_score, roc_auc_score
from tensorflow.keras.callbacks import EarlyStopping
import seaborn as sns
from sklearn.preprocessing import label_binarize
```

```python
# Define a simple classifier with Input layer
inputs = Input(shape=features.shape[1:])
x = GlobalAveragePooling2D()(inputs)
x = Dense(1024, activation='relu')(x)
x = Dropout(0.5)(x)  # Add Dropout
x = Dense(512, activation='relu')(x)
x = Dropout(0.5)(x)  # Add Dropout
outputs = Dense(n_clusters, activation='softmax')(x)

classifier = Model(inputs, outputs)
```

```python
# Compile the model
classifier.compile(optimizer=Adam(learning_rate=0.001), loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
# Early stopping callback
early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
```

```python
# Train the model with the pseudo-labels
history = classifier.fit(
    features,
    clusters,
    epochs=20,
    batch_size=16,
    validation_split=0.2,
    callbacks=[early_stopping],
    verbose=1  #Set verbose to 1 to print training progress
)

```

```python
# Evaluate metrics on validation data
val_features = features[int(len(features) * 0.8):]
val_clusters = clusters[int(len(clusters) * 0.8):]
predictions = classifier.predict(val_features)
predicted_labels = np.argmax(predictions, axis=1)
```

```python
# Compute Accuracy
accuracy = accuracy_score(val_clusters, predicted_labels)
print(f'Accuracy: {accuracy:.4f}')
```

```python
# Compute Precision, Recall, F1 Score
precision = precision_score(val_clusters, predicted_labels, average='weighted')
recall = recall_score(val_clusters, predicted_labels, average='weighted')
f1 = f1_score(val_clusters, predicted_labels, average='weighted')

print(f'Precision: {precision:.4f}')
print(f'Recall: {recall:.4f}')
print(f'F1 Score: {f1:.4f}')
```

```python
# Compute the confusion matrix
cm = confusion_matrix(val_clusters, predicted_labels)
```

```python
# Plot the confusion matrix
plt.figure(figsize=(10, 7))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=np.arange(n_clusters), yticklabels=np.arange(n_clusters))
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.title('Confusion Matrix')
plt.show()
```

```python
# Compute AUC-ROC for each class
# Binarize the output
val_clusters_bin = label_binarize(val_clusters, classes=np.arange(n_clusters))
predictions_bin = predictions  # Predicted probabilities are already in the required format
```

```python
# Compute AUC-ROC
roc_auc = roc_auc_score(val_clusters_bin, predictions_bin, average='weighted', multi_class='ovr')
print(f'AUC-ROC: {roc_auc:.4f}')
```

```python
# Class-specific Precision and Recall
class_names = ['Normal', 'Osteopenia'] 
```

```python
normal_index = 0  # Index for 'Normal'
osteopenia_index = 1  # Index for 'Osteopenia'
```

```python
precision_normal = precision_score(val_clusters, predicted_labels, labels=[normal_index], average='macro')
recall_normal = recall_score(val_clusters, predicted_labels, labels=[normal_index], average='macro')
precision_osteopenia = precision_score(val_clusters, predicted_labels, labels=[osteopenia_index], average='macro')
recall_osteopenia = recall_score(val_clusters, predicted_labels, labels=[osteopenia_index], average='macro')
```

```python
print(f'Precision (Normal): {precision_normal:.4f}')
print(f'Recall (Normal): {recall_normal:.4f}')
print(f'Precision (Osteopenia): {precision_osteopenia:.4f}')
print(f'Recall (Osteopenia): {recall_osteopenia:.4f}')
```

# Using the Classifier to Detect Conditions

```python
# Load the classifier model
from tensorflow.keras.models import load_model

classifier = load_model('bone_density_classifier.keras')
```

```python
def estimate_condition(image):
    feature = model.predict(np.expand_dims(image, axis=0))  # Extract features
    prediction = classifier.predict(feature)  # Predict the class
    class_idx = np.argmax(prediction)
    
    if class_idx == 0:
        return 'Condition: Normal', 'T-score ≥ -1'
    elif class_idx == 1:
        return 'Condition: Osteopenia', '-2.5 < T-score < -1'
    else:
        return 'Condition: Osteoporosis', 'T-score ≤ -2.5'
```

```python
# Example usage
example_image_path = r"C:\Harsh\Bone Density Internship\Dataset\dicom_dir\ID_0071_AGE_0065_CONTRAST_0_CT.dcm"
example_image = load_dicom_image(example_image_path)
condition, t_score = estimate_condition(example_image)
print(f'{condition}, Estimated T-score: {t_score}')
```
